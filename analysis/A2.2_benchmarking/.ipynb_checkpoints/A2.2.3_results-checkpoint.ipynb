# 1. Raw, main metrics

```python
# labelling

pp = pps[1]
lresult1 = {}
for method in LR_methods:
    scores = {}
    for metric in metrics:
        score = []
        for dataset in ldatasets:
            x = f"{home}{'_'.join([dataset,method,pp,metric])}.npy"
            if ('maxi' not in x) or ('PTS' not in x): # this dataset doesn't exist
                score.append(np.load(x).mean())
        scores[metric] = score
    lresult1[method] = scores
            
    
```

```python
# splicing

pp = pps[1]
sresult1 = {}
for method in SR_methods:
    scores = {}
    for metric in metrics:
        score = []
        for dataset in sdatasets:
            x = f"{home}{'_'.join([dataset,method,pp,metric])}.npy"
            if ('maxi' not in x) or ('PTS' not in x): # this dataset doesn't exist
                score.append(np.load(x).mean())
        scores[metric] = score
    sresult1[method] = scores
            
    
```

```python
proper_label_dict = {
    'velvet':'VelvetVAE',
    'svelvet':'SvelvetVAE',
    'velvetSDE':'VelvetSDE',
    'dynamo':'Dynamo',
    'scvelo':'scVelo',
    'unitvelo':'UniTVelo',
    'velovi':'Velovi',
    'kappavelo':'Îº-velo',
    'latentvelo':'Latentvelo',
    'pyrovelocity':'Pyrovelocity',
    'cell2fate':'Cell2Fate',
    'Positive Control':'+ Control',
    'deepvelo':'DeepVelo',
    'velovae':'VeloVAE'
}
```

```python
result1 = {**lresult1, **sresult1}

new_order = ['velvet', 'svelvet', 'velvetSDE', 'dynamo', 'scvelo', 'unitvelo', 'velovi', 'kappavelo', 'latentvelo', 'pyrovelocity', 'cell2fate']

result_reordered = {key: result1[key] for key in new_order}


data = []
for group, subdict in result_reordered.items():
    for subgroup, values in subdict.items():
        for value in values:
            data.append({'Group': group, 'Subgroup': subgroup, 'Value': value})

df = pd.DataFrame(data)
subgroups = set(df['Subgroup'])

for sg in ['CRS','PTS','CBD']:
    pc = np.load(home+'positive_control.npy')
    pc2 = np.load(home+'CBD_PC.npy')
    if sg=='CBD':
        pcdf_new = pd.DataFrame(
            {'Group':['Positive Control']*len(pc2),
             'Subgroup':[sg]*len(pc2),
             'Value':pc2}
        )
    else:
        pcdf_new = pd.DataFrame(
            {'Group':['Positive Control']*len(pc),
             'Subgroup':[sg]*len(pc),
             'Value':pc}
        )
    df = pd.concat([df, pcdf_new], ignore_index=False)


l, s, g = '#FF9999','#99CCFF','lightgrey'
colors = [l,s,l,l,s,s,s,s,s,s,s,g]
groups = df['Group'].unique()
color_map = dict(zip(groups, colors))

for subgroup in subgroups:
    plt.figure(figsize=(3,5), dpi=300)
    sns.violinplot(
        y='Group', 
        x='Value',
        data=df[df['Subgroup'] == subgroup], 
        scale='width', 
        palette=color_map,
        zorder=4,
        cut=0
    )
    plt.title(f'{subgroup} score', fontsize=25)
    for label in plt.gca().get_yticklabels():
        if 'velvet' in label.get_text():
            label.set_weight('bold')
    labels = [label.get_text() for label in plt.gca().get_yticklabels()]
    proper_labels = [proper_label_dict[l] for l in labels]
    plt.gca().set_yticklabels(proper_labels, fontsize=18)
    for i in range(len(labels)):
        plt.axhline(y=i,zorder=0, color='lightgrey')
    if subgroup != 'CBD':
        plt.axvline(x=np.median(pc), zorder=0, color='lightgrey', linestyle='--')
    else:
        plt.axvline(x=np.median(pc2), zorder=0, color='lightgrey', linestyle='--')

    if subgroup != 'PTS':
        plt.gca().set_yticklabels([])
    plt.xticks(fontsize=12)
    plt.gca().set_ylabel('')
    plt.gca().set_xlabel('')
    
    for fig in plt.get_fignums():
        fig = plt.figure(fig)
        fig.patch.set_facecolor('none')
        for ax in fig.get_axes():
            ax.set_facecolor('white')
    plt.savefig(f"../../figures/2.2.3_benchmark_score_raw_{subgroup}.png", bbox_inches='tight')
    plt.show()
```

# 6. component scores

```python
cond = [
    'NoC_NoG',
    'WiC_NoG',
    'NoC_WiG',
]
```

```python
cresult = {}
for c in cond:
    met = {}
    for metric in metrics:
        sub = []
        for dataset in ldatasets:
            file = f'{dataset}_velvet_{c}_{metric}.npy'
            if ('maxi' not in file) or ('PTS' not in file): # this dataset doesn't exist
                x = np.load(home+file).mean()
            sub.append(x)
        met[metric] = sub
    cresult[c] = met
        
met = {}
for metric in metrics:
    sub = []
    for dataset in ldatasets:
        file = f'{dataset}_velvet_RAW_{metric}.npy'
        if ('maxi' not in file) or ('PTS' not in file): # this dataset doesn't exist
            x = np.load(home+file).mean()
        sub.append(x)
    met[metric] = sub
cresult['WiC_WiG'] = met        


data = []
for group, subdict in cresult.items():
    for subgroup, values in subdict.items():
        for value in values:
            data.append({'Group': group, 'Subgroup': subgroup, 'Value': value})
            
df = pd.DataFrame(data)
subgroups = set(df['Subgroup'])
```

```python
dfsub = df.loc[[d in ['NoC_WiG','WiC_WiG'] for d in df.Group]]

dfsub['Group'] = [g.replace(
    'WiC_WiG','+NC'
).replace(
    'NoC_WiG','-NC'
) for g in dfsub.Group]

plt.figure(figsize=(3,3), dpi=300)
sns.violinplot(x='Subgroup', y='Value', hue='Group', data=dfsub, scale='width', palette="colorblind",)
plt.legend(loc='upper left')
plt.ylabel("Score", fontsize=15)
plt.xlabel("Metric", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

for fig in plt.get_fignums():
    fig = plt.figure(fig)
    fig.patch.set_facecolor('none')
    for ax in fig.get_axes():
        ax.set_facecolor('white')
plt.savefig(f"../../figures/2.2.3_component_scores.png", bbox_inches='tight')

plt.show()
```
